# Visualización de datos

El dataset que escogí es el data set Titan

```python
import pandas as pd

# Ruta relativa al dataset Titan
titan_path = "../Entrega 1/Data Set 2/Titan.csv"

# Cargar el dataset utilizando Pandas
titan_df = pd.read_csv(titan_path)

# Visualizar las primeras filas del dataset
titan_df.head()
```

## Matplotlib


```python
import matplotlib.pyplot as plt


```

```python
# Convertir la columna 'Date' a tipo datetime
titan_df['Date'] = pd.to_datetime(titan_df['Date'])

# Agrupar por mes y calcular el precio promedio de cierre
monthly_avg_close = titan_df.groupby(pd.Grouper(key='Date', freq='M'))['Close'].mean()

# Gráfico de líneas: Precio promedio de cierre por mes
plt.plot(monthly_avg_close.index, monthly_avg_close.values)
plt.xlabel('Fecha')
plt.ylabel('Precio promedio de cierre')
plt.title('Precio promedio de cierre del stock de Titan por mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Agrupar el dataset por mes y calcular el volumen de transacciones promedio
titan_df['Date'] = pd.to_datetime(titan_df['Date'])
monthly_avg_volume = titan_df.groupby(titan_df['Date'].dt.to_period('Y'))['Volume'].mean()

# Gráfico de barras: Volumen de transacciones promedio por mes
monthly_avg_volume.plot(kind='bar')
plt.xlabel('Mes')
plt.ylabel('Volumen de transacciones promedio')
plt.title('Volumen de transacciones promedio por mes del stock de Titan')

# Interpretación: Este gráfico muestra el volumen de transacciones promedio agrupado por mes. Podemos comparar visualmente el volumen promedio entre los diferentes meses para identificar posibles patrones o tendencias mensuales.

plt.show()
```

```python

import random

# Tamaño de la muestra (ajusta según tus necesidades)
sample_size = 1000

# Obtener una muestra aleatoria de puntos
random_indices = random.sample(range(len(titan_df)), sample_size)
sampled_data = titan_df.iloc[random_indices]

# Gráfico de dispersión con muestreo aleatorio: Precio de apertura vs. Precio de cierre
plt.scatter(sampled_data['Open'], sampled_data['Close'],alpha=0.3)
plt.xlabel('Precio de apertura')
plt.ylabel('Precio de cierre')
plt.title('Relación entre el precio de apertura y el precio de cierre (Muestreo Aleatorio)')
plt.show()
```

Gráfico de líneas: Precio promedio de cierre por mes
Este gráfico muestra la tendencia del precio promedio de cierre del stock de Titan a lo largo del tiempo. Podemos observar si hay cambios significativos en el precio promedio de cierre en diferentes meses.

Gráfico de barras: Volumen de transacciones promedio por mes
Este gráfico muestra el volumen promedio de transacciones agrupado por mes. Nos permite comparar visualmente el volumen promedio entre los diferentes meses y detectar posibles patrones o tendencias mensuales.

Gráfico de dispersión: Relación entre el precio de apertura y el precio de cierre (Muestreo Aleatorio)
Este gráfico de dispersión muestra la relación entre el precio de apertura y el precio de cierre del stock de Titan. Al utilizar un muestreo aleatorio, se selecciona un subconjunto de puntos para representar la relación de manera más comprensible. Podemos analizar la dispersión y la dirección de la relación entre estas dos variables.

# Seaborn

```python
import seaborn as sns


```

```python

# Gráfico de caja y bigotes: Distribución del precio de cierre por día de la semana
sns.boxplot(data=titan_df, x=titan_df['Date'].dt.weekday, y='Close')
plt.xlabel('Día de la semana')
plt.ylabel('Precio de cierre')
plt.title('Distribución del precio de cierre por día de la semana')
plt.show()
```

```python

# Gráfico de violín: Distribución del volumen de transacciones por mes
sns.violinplot(data=titan_df, x=titan_df['Date'].dt.month, y='Volume')
plt.xlabel('Mes')
plt.ylabel('Volumen de transacciones')
plt.title('Distribución del volumen de transacciones por mes')
plt.show()
```

```python

# Agregar el dataset por mes y calcular el precio promedio de cierre
monthly_avg_close = titan_df.groupby(titan_df['Date'].dt.to_period('Y'))['Close'].mean().reset_index()

# Gráfico de barras: Precio promedio de cierre por mes
sns.barplot(data=monthly_avg_close, x='Date', y='Close')
plt.xlabel('Mes')
plt.ylabel('Precio promedio de cierre')
plt.title('Precio promedio de cierre por mes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Gráfico de caja y bigotes: Distribución del precio de cierre por día de la semana
Este gráfico de caja y bigotes muestra la distribución del precio de cierre del stock de Titan para cada día de la semana. Podemos observar la mediana, los cuartiles y los valores atípicos, lo que nos brinda información sobre la variabilidad y los posibles patrones en la distribución del precio de cierre según el día de la semana.

Gráfico de violín: Distribución del volumen de transacciones por mes
Este gráfico de violín muestra la distribución del volumen de transacciones del stock de Titan para cada mes. Podemos observar la forma de la distribución y comparar la densidad en diferentes meses, lo que nos proporciona información sobre la variabilidad y los posibles patrones en el volumen de transacciones mensuales.

Gráfico de barras: Precio promedio de cierre por mes
Este gráfico de barras muestra el precio promedio de cierre del stock de Titan por mes. Nos permite comparar visualmente el precio promedio entre los diferentes meses y detectar posibles variaciones significativas en el precio de cierre a lo largo del tiempo.
